```python
import pandas as pd
import requests
import re
```

```python
# 同花顺网页数据

headers = {"Accept": "application/json, text/javascript, */*; q=0.01",
           "Referer": "http://www.10jqka.com.cn/",
           "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/86.0.4240.111 Safari/537.36"}


# 获取前十页内容
url = "http://data.10jqka.com.cn/ajax/yjyg/agentSource/static1610175062/date/2020-12-31/board/ALL/field/enddate/order/desc/page/{}/ajax/1/free/1/"
result = ''


for i in range (0,11):
    result += requests.get(url.format(i), headers = headers).text

result
```

```python
# 获取序号
rank = re.findall(r'<td>(\d+)</td>', result)
rank
```

```python
# 获取股票代码
stockcode = r'class="stockCode">(\d+)</a>'
stock_codes = re.findall(stockcode, result)
stock_codes
```

```python
# 获取股票名称
stock_names = re.findall('class="J_showCanvas">(.+)</a>', result)
stock_names
```

```python
# 业绩预告类型
forecast_types = re.findall("<span>(.+)</span>", result)
forecast_types
```

```python
# 业绩预告摘要
summary = re.findall(r'target="_blank">(.*)</a>', result)
summary
```

```python
# 净利润变动幅度(%)
# 净利润变动幅度里的标签是不固定的，有时是<td class =" c-rise">， 有时是<td class =" c-fall">
# 这个数据里面有空值，空值表示为'-'
percentage_change_regex = '''>
            	(.*)            </td>
            <td class='''


percentage_change = re.findall(percentage_change_regex, result)
percentage_change
```

```python
print(len(percentage_change))
```

```python
# 上年同期净利润

'''</td>
            <td class="tc cur">'''

net_profit_regex = '>(.*)</td>\n            <td class="tc cur">'
net_profit = re.findall(net_profit_regex, result)
net_profit
```

```python
len(net_profit)
```

```python
# 获取公布日期
date = r'<td class="tc cur">(\d+-\d+-\d+)</td>'
publish_dates = re.findall(date, result)
publish_dates
```

```python
cols = {"序号": rank, "股票代码": stock_codes,
        "股票名称": stock_names, "发布日期": publish_dates,
       "业绩预告类型": forecast_types, "业绩预告摘要": summary,
       "净利润变动幅度(%)": percentage_change,"上年同期净利润(元)": net_profit}
data = pd.DataFrame(cols)
data.head()
```

```python
import matplotlib.pyplot as plt

font={
    "family":"SimHei",
    "size":15
}
plt.rc('font',**font)
```

```python
# 查看业绩预告类型的分布
performance_tb = data.groupby("业绩预告类型", as_index = False)["序号"].agg({"数量": "count"}).sort_values(by = "数量", ascending = False)
performance_tb
```

```python
plt.figure(figsize = (10,5))
plt.title("500家公司业绩预告类型分布", size = 16)
plt.bar(x = "业绩预告类型", height = "数量", data = performance_tb)
plt.xticks(rotation = 45)
plt.show()
```

```python
# 我们认为 '业绩预降', '业绩大幅上升', '业绩大幅下降', '业绩预盈', '业绩预增', '预计扭亏'   是盈利的
# 认为   '预计减亏', '不确定','业绩预亏', '预计增亏', '预计续亏'  是亏损的

def if_profit(x):
    if x in ['业绩预降', '业绩大幅上升', '业绩大幅下降', '业绩预盈', '业绩预增', '预计扭亏']:
        return 1
    if x in ['预计减亏', '不确定','业绩预亏', '预计增亏', '预计续亏']:
        return -1
    return 0
```

```python
data['是否盈利'] = data.apply(lambda x : if_profit(x['业绩预告类型']),axis=1)
data.head()
```

```python
# 盈利的数量
gain = (data['是否盈利'] == 1).sum()
gain
```

```python
# 亏损的数量
loss = (data['是否盈利'] == -1).sum()
loss
```

```python
plt.pie([gain, loss], labels=['盈利','亏损'], autopct='%.2f%%')

plt.title('250家上市公司盈利和亏损情况占比')
plt.show()
```

```python
# 查看净利润变动幅度的缺失值数量
is_missing = sum(data["净利润变动幅度(%)"] == "-")
not_missing = sum(data["净利润变动幅度(%)"] != "-")
print("净利润变动幅度值缺失个数：", is_missing,"\n净利润变动幅度值一共有：",is_missing+not_missing,"个")
```

```python
# 用饼图查看
plt.title("净利润变动幅度值缺失占比", size = 16)
plt.pie([is_missing, not_missing],labels=['缺失','未缺失'], autopct='%.2f%%')
plt.show()
```
